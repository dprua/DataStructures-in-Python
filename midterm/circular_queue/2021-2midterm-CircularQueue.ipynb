__Solution__:

````python
#%%writefile queueCircular.py

class QueueCircular:
    def __init__(self, size):
        self.items = [None] * size
        self.MAX = size
        self.front = -1
        self.back = -1
        
    def enqueue(self, item): # if not full
        if self.is_full(): # full 이라면 The circular queue is full[front,back]
            sen = "The circular queue is full[{0}:{1}].".format(self.front,self.back)
            print(sen)
            return sen
        if self.front == -1 and self.back == -1:
            self.front = 0
            self.back = 0
            self.items[self.back] = item
        else:
            self.back = (self.back+1) % self.MAX
            self.items[self.back] = item
            

    def dequeue(self): # if not empty
        if self.is_empty():
            sen = "The circular queue is empty[{0}:{1}].".format(self.front,self.back)
            print(sen)
            return False
        else:
            val = self.items[self.front]
            if self.front == self.back:
                self.front = -1
                self.back = -1
            else:
                self.front = (self.front+1) % self.MAX
            return val
    
    def is_empty(self):
        return self.front == -1

    def is_full(self):
        if self.front == 0 and self.back == self.MAX -1:
            return True
        elif self.front == self.back + 1:
            return True
        else:
            return False
    
    def size(self):
        if self.is_empty():
            return 0
        else:
            count = 1
            f = self.front
            b = self.back
            while f != b:
                f = (f+1) % self.MAX
                count += 1
            return count

    def __repr__(self):
        pstr = 'QueueCircular(['
        if self.front == -1:
            return pstr + '])'
        f = self.front
        ran = self.size()
        for i in range(ran):
            if i == self.size() - 1:
                pstr += str(self.items[f])
            else:
                pstr += str(self.items[f])
                pstr += ', '
            f = (f+1) % self.MAX
        return pstr + '])'
````

### Step 1 (2.0p)
다음은 `QueueCircular` 클래스를 수정하여, 파이썬 `deque`의 `maxlen` 특성을 일부분 살릴 수 있는 `QueueMaxlen` 클래스를 작성하십시오. 

- `QueueMaxlen`는 항상 `maxlen`을 유지하며, `enqueue()`메소드에서 overflow가 일어나려고 할 때, `front`에 있는 가장 오래된 요소를 삭제하여 입력되는 최근의 요소들을 항상 입력되도록 합니다. 이러한 큐에서는 overflow가 일어나지 않습니다. 
- Default `maxlen = 3` 으로 설정합니다. 


__Sample Run:__
```
if __name__ == '__main__':
    four = QueueMaxlen(maxlen = 4)
    print(four)
    for i in range(5, 10):        # Discard 5      
        four.enqueue(i)
    print(four)
    four.enqueue(1)
    four.enqueue(2)
    print(four)
    four.dequeue()
    print(four)
    print(four.maxlen)
```

__Expected Output:__
```
QueueMaxlen([])
QueueMaxlen([6, 7, 8, 9])
QueueMaxlen([8, 9, 1, 2])
QueueMaxlen([9, 1, 2])
4
```

````python
#%%writefile queueMaxlen.py

class QueueMaxlen:
    def __init__(self, maxlen = 3):
        self.maxlen = maxlen
        self.items = [None] * self.maxlen
        self.front = -1
        self.back = -1
        
    def enqueue(self, item): 
        if self.front == -1 and self.back == -1:
            self.front = 0
            self.back = 0
            self.items[self.back] = item
        else:
            if self.is_full():
                self.front = (self.front+1) % self.maxlen
                self.back = (self.back+1) % self.maxlen
                if type(item) == str:
                    val = "\"" + item + "\""
                    item = val
                self.items[self.back] = item 
            else: 
                self.back = (self.back+1) % self.maxlen
                if type(item) == str:
                    val = "\"" + item + "\""
                    item = val
                self.items[self.back] = item

    def dequeue(self): # if not empty
        if self.is_empty():
            sen = "The circular queue is empty[{0},{1}].".format(self.front,self.back)
            print(sen)
            return False
        else:
            val = self.items[self.front]
            if self.front == self.back:
                self.front = -1
                self.back = -1
            else:
                self.front = (self.front+1) % self.maxlen
            return val

    def is_empty(self):
        return self.front == -1

    def is_full(self):
        if self.front == 0 and self.back == self.maxlen -1:
            return True
        elif self.front == self.back + 1:
            return True
        else:
            return False
    
    def size(self):
        if self.is_empty():
            return 0
        else:
            count = 1
            f = self.front
            b = self.back
            while f != b:
                f = (f+1) % self.maxlen
                count += 1
            return count

    def __repr__(self):            
        pstr = 'QueueMaxlen(['
        if self.front == -1:
            return pstr + '])'
        f = self.front
        ran = self.size()
        for i in range(ran):
            if i == self.size() - 1:
                pstr += str(self.items[f])
            else:
                pstr += str(self.items[f])
                pstr += ', '
            f = (f+1) % self.maxlen
        return pstr + '])'
````

## Problem 3 (1.0p) - Subclass `QueueMaxlenSub`

다음은 `QueueCircular` 클래스를 Superclass로 설정하고, `QueueMaxlenSub` 클래스를 Subclass로 만드십시오. 그리고, 가능한 OOP의 inhertance특성을 모두 활용하고, 반드시 필요한 메소드들만 여기서 구현합니다.  모든 기능은 Problem 2와 같고, 테스트 스크립트도 같지만(물론, 클래스 이름은 다름), 코딩은 좀 더 간단해지는 것을 확인하십시오.  

- 여기서 클래스 이름은 `QueueMaxlenSub`인 것에 유의하십시오. 


````python
# your subclass 'QueueMaxlenSub' here

from abc import ABC, abstractmethod

class QueueCircular(ABC):
    def __init__(self, size):
        self.items = [None] * size
        self.MAX = size
        self.front = -1
        self.back = -1
    @abstractmethod
    def __repr__(self):
        pass
    
    @abstractmethod
    def enqueue(self, item): # if not full
        pass
            
    @abstractmethod
    def dequeue(self): # if not empty
        pass
    
    def is_empty(self):
        return self.front == -1

    def is_full(self):
        if self.front == 0 and self.back == self.MAX -1:
            return True
        elif self.front == self.back + 1:
            return True
        else:
            return False
    
    def size(self):
        if self.is_empty():
            return 0
        else:
            count = 1
            f = self.front
            b = self.back
            while f != b:
                f = (f+1) % self.MAX
                count += 1
            return count   
    
class QueueMaxlenSub(QueueCircular):
    def __init__(self, maxlen = 3):
        QueueCircular.__init__(self,maxlen)
        self.maxlen = maxlen
        
    def enqueue(self, item): 
        if self.front == -1 and self.back == -1:
            self.front = 0
            self.back = 0
            self.items[self.back] = item
        else:
            if self.is_full():
                self.front = (self.front+1) % self.maxlen
                self.back = (self.back+1) % self.maxlen
                if type(item) == str:
                    val = "\"" + item + "\""
                    item = val
                self.items[self.back] = item 
            else: 
                self.back = (self.back+1) % self.maxlen
                if type(item) == str:
                    val = "\"" + item + "\""
                    item = val
                self.items[self.back] = item

    def dequeue(self): # if not empty
        if self.is_empty():
            sen = "The circular queue is empty[{0},{1}].".format(self.front,self.back)
            print(sen)
            return False
        else:
            val = self.items[self.front]
            if self.front == self.back:
                self.front = -1
                self.back = -1
            else:
                self.front = (self.front+1) % self.maxlen
            return val
        
    def __repr__(self):            
        pstr = 'QueueMaxlenSub(['
        if self.front == -1:
            return pstr + '])'
        f = self.front
        ran = self.size()
        for i in range(ran):
            if i == self.size() - 1:
                pstr += str(self.items[f])
            else:
                pstr += str(self.items[f])
                pstr += ', '
            f = (f+1) % self.maxlen
        return pstr + '])'
````
